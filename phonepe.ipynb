```python
#extract aggregate transaction data from json file

def agg_trans_data():
    path1='phonepe/data/aggregated/transaction/country/india/state/'
    agg_trans_state_list=os.listdir(path1)

    tran_clm={'state':[],'year':[],'Quater':[],'Transaction_type':[], 'Transaction_count':[], 'Transaction_amount':[]}

    for state in agg_trans_state_list:
        p_state=path1 + state + "/"
        agg_year=os.listdir(p_state)
        for year in agg_year:
            p_year=p_state+ year+"/"
            agg_Quater=os.listdir(p_year)
            for quater in agg_Quater:
                p_quater=p_year + quater
                data=open(p_quater,'r')
                d=json.load(data)
                for i in d['data']['transactionData']:
                    name=i['name']
                    count=i['paymentInstruments'][0]['count']
                    amount=i['paymentInstruments'][0]['amount']
                    tran_clm['Transaction_type'].append(name)
                    tran_clm['Transaction_count'].append(count)
                    tran_clm['Transaction_amount'].append(amount)
                    tran_clm['state'].append(state)
                    tran_clm['year'].append(year)
                    tran_clm['Quater'].append(int(quater.strip('.json')))

    df=pd.DataFrame(tran_clm)
    df['state']=df['state'].str.replace("-"," ").str.title()
    df['state']=df['state'].str.replace("Andaman & Nicobar Islands","Andaman & Nicobar")
    df['state']=df['state'].str.replace('Dadra & Nagar Haveli & Daman & Diu','Dadra and Nagar Haveli and Daman and Diu')
        
    return(df)
```

```python
#extract map user data from json file

def map_user_data():
    path4="phonepe/data/map/user/hover/country/india/state/"
    map_user_state_list=os.listdir(path4)

    user_clm={'state':[],'District':[],'year':[],'Quater':[],'Registered_Users':[],'App_Opens':[]}

    for state in map_user_state_list:
        p_state=path4 + state + "/"
        agg_year=os.listdir(p_state)
        for year in agg_year:
            p_year=p_state+ year+"/"
            agg_Quater=os.listdir(p_year)
            for quater in agg_Quater:
                p_quater=p_year + quater
                data=open(p_quater,'r')
                d=json.load(data)
                for district,data_dist in d['data']['hoverData'].items():
                    registeredUsers=data_dist['registeredUsers']
                    appopens=data_dist['appOpens']
                    user_clm['District'].append(district)
                    user_clm['Registered_Users'].append(registeredUsers)
                    user_clm['App_Opens'].append(appopens)
                    user_clm['state'].append(state)
                    user_clm['year'].append(year)
                    user_clm['Quater'].append(int(quater.strip('.json')))

    df=pd.DataFrame(user_clm)
    df['state']=df['state'].str.replace("-"," ").str.title()
    df['state']=df['state'].str.replace("Andaman & Nicobar Islands","Andaman & Nicobar")
    df['state']=df['state'].str.replace('Dadra & Nagar Haveli & Daman & Diu','Dadra and Nagar Haveli and Daman and Diu')
    df['District']=df['District'].str.replace('district','').str.title()
    return(df)
```

```python
#mysql connetor to connect with python
import mysql.connector
from sqlalchemy import create_engine

mydb = mysql.connector.connect(host="localhost",user="root",password="")
mycursor = mydb.cursor(buffered=True,)

#create a engine to insert data value
engine = create_engine("mysql+mysqlconnector://root:@localhost/phonepe") 

#create database and use for table creation
mycursor.execute('create database if not exists phonepe')
mycursor.execute('use phonepe')
```

```python
# execute sql query to create a table and inserted value with sql alchemy engine

mycursor.execute('''create table if not exists agg_transaction(State VARCHAR(100),
                Year INT,Quater INT,Transacion_type VARCHAR(300),Transacion_count BIGINT,
                Transacion_amount FLOAT)''')

mycursor.execute('''create table if not exists agg_user(State VARCHAR(100),
                Year INT,Quater INT,User_brand VARCHAR(50),User_count BIGINT,
                User_percentage FLOAT)''')

mycursor.execute('''create table if not exists map_transaction(State VARCHAR(100),District VARCHAR(100),
                Year INT,Quater INT,Transacion_count BIGINT,Transacion_amount FLOAT )''')

mycursor.execute('''create table if not exists map_user(State VARCHAR(100),District VARCHAR(100),
                Year INT,Quater INT,Registered_Users BIGINT,App_Opens BIGINT)''')

mycursor.execute('''create table if not exists top_transaction(State VARCHAR(100),Pincode INT,
                Year INT,Quater INT,Transacion_count BIGINT,Transacion_amount FLOAT)''')

mycursor.execute('''create table if not exists top_user(State VARCHAR(100),Pincode INT,
                Year INT,Quater INT,Registered_Users BIGINT)''')

agg_trans_data().to_sql('agg_transaction',engine,if_exists='replace',index=False)
agg_user_data().to_sql('agg_user',engine,if_exists='replace',index=False)
map_trans_data().to_sql('map_transaction',engine,if_exists='replace',index=False)
map_user_data().to_sql('map_user',engine,if_exists='replace',index=False)
top_trans_data().to_sql('top_transaction',engine,if_exists='replace',index=False)
top_user_data().to_sql('top_user',engine,if_exists='replace',index=False)
mydb.commit()
```
